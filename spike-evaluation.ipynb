## Analyze spikes with bowtie 1

```python
! bowtie --version
```

```python
def parse_bowtie1(result):
    parsed = {'total': None,
              'hits': None,
              'failed': None,
              'multi': None}
    
    terms = [ ('reads processed: ([\d]+)', 'total'),
              ('one reported alignment: ([\d]+)', 'hits'),
              ('failed to align: ([\d]+)', 'failed'),
              ('alignments suppressed due to -m: ([\d]+)', 'multi')]
    for line in result:
        for query, variable in terms:
            match = re.search(query, line)
            if match:
                parsed[variable] = int(match.group(1))
    return parsed
```

```python
 def bowtie(index_name, mismatches, read_length):
    """Using some ipython features run bowtie from python
    
    -v specifies the allowed number of mismatches in a range 0-3.
    -k (return upto N matches)
    -m (suppress reads with more than ths number of hits)
    
    Using -m 1 should require all reads to be unique.
    """
    result = ! bowtie indexes/{index_name}  -v {mismatches} -m 1 --best --strata -f \
                      spikes-{read_length}.fa \
                      --un {mismatches}-mm/unmapped-{index_name}-{read_length}.fa \
                      --max {mismatches}-mm/repeat-{index_name}-{read_length}.fa \
                      {mismatches}-mm/hits-{index_name}-{read_length}.txt
    return parse_bowtie1(result)
                
```

```python
def process_bowtie1():
    # uncomment for testing.
    #indexes = ['spikes']
    if 'bowtie1' in cache:
        return cache['bowtie1']
    processed = []
    for index in indexes:
        for mismatch in mismatches:
            for read_length in read_lengths:
                properties = {
                   'index': index,
                   'mismatch': mismatch,
                   'read_length': read_length}
                result = bowtie(index, mismatch, read_length)
                properties.update(result)
                processed.append(properties)
    df =  pandas.DataFrame(processed, columns=['index', 'mismatch', 'read_length', 'total', 'hits', 'multi', 'failed', ])
    cache['bowtie1'] = df
    return df
```

```python
process_bowtie1()
```

## Analyze Bowtie2

```python
! bowtie2 --version
```

```python
def parse_bowtie2(result):
    parsed = {'total': None,
              'hits': None,
              'failed': None,
              'multi': None}
    
    terms = [ ('(?P<total>[\d]+) reads;', 'total'),
              ('(?P<failed>[\d]+).*aligned 0 times', 'failed'),
              ('(?P<unique>[\d]+).*aligned exactly 1 time', 'hits'),
              ('(?P<multi>[\d]+).*aligned >1 times', 'multi'),
              ]
    for line in result:
        for pattern, variable in terms:
            match = re.search(pattern, line)
            if match:
                parsed[variable] = int(match.group(1))
    return parsed
```

```python
def bowtie2(index_name, read_length):
    """Run bowtie 2, using gravely lab settings 
    """
    result = ! bowtie2 -k 5 -f -x indexes2/{index_name} -U spikes-{read_length}.fa \
                       --un 0-mm/unmapped2-{index_name}-{read_length}.fa \
                       -S 0-mm/hits-{index_name}-{read_length}.sam
    return parse_bowtie2(result)
```

```python
def process_bowtie2():
    # uncomment for testing.
    #indexes = ['spikes']
    if 'bowtie2' in cache:
        return cache['bowtie2']
    processed = []
    for index in indexes:
        for read_length in read_lengths:
            properties = {
               'index': index,
               'read_length': read_length}
            result = bowtie2(index, read_length)
            properties.update(result)
            processed.append(properties)
    df =  pandas.DataFrame(processed, columns=['index', 'read_length', 'total', 'hits', 'multi', 'failed', ])
    cache['bowtie2'] = df
    return df
```

```python
process_bowtie2()
```
